### Authenticate

```python
def authenticate():
    '''
    Authenticate
    Authenticates your credentials and creates a client.
    '''
    load_dotenv()
    subscription_key = os.environ["VISION_KEY"]
    endpoint = os.environ["VISION_ENDPOINT"]
    credentials = CognitiveServicesCredentials(subscription_key)
    computervision_client = ComputerVisionClient(endpoint, credentials)
    '''
    END - Authenticate
    '''
    return computervision_client
```

```python
#setting tags for bike items
computervision_client = authenticate()
bycicle_obj = ["bicycle", "bike", "cycle", "bicycle wheel", "bicycle tire", "bicycle frame", "bicycle handlebar",
               "bicycle seat", "bicycle pedal", "bicycle chain", "bicycle saddle", "bicycle fork"]
```

## Getting classes for each image
This will return the tags in that image

```python

def get_classes(image_path):
    with open(image_path, "rb") as image_stream:
        image_analysis = computervision_client.analyze_image_in_stream(image_stream,
                                                                       visual_features=[VisualFeatureTypes.tags,
                                                                                        VisualFeatureTypes.objects])
    print("tags")
    tags = {}
    for tag in image_analysis.tags:
        if tag.name not in tags:
            tags[tag.name] = [tag.confidence]
        else:
            tags[tag.name].append(tag.confidence)
    return tags
```

## Getting tags and objects
This will return the tags and 
            the objects found in the image given based of the object searched . Also it will give the location of them in the image

```python
def get_predicted_class(image_path, object_searched):
    with open(image_path, "rb") as image_stream:
        image_analysis = computervision_client.analyze_image_in_stream(image_stream,
                                                                       visual_features=[VisualFeatureTypes.tags,
                                                                                        VisualFeatureTypes.objects])
    # print("tags")
    tags = {}
    for tag in image_analysis.tags:
        if tag.name in object_searched:
            if tag.name not in tags:
                tags[tag.name] = [tag.confidence]
            else:
                tags[tag.name].append(tag.confidence)

    # print("objects")
    objects = {}
    for ob in image_analysis.objects:
        if ob.object_property in object_searched:
            location = [ob.rectangle.x, ob.rectangle.y, ob.rectangle.w + ob.rectangle.x,
                        ob.rectangle.h + ob.rectangle.y]
            if ob.object_property not in objects:
                objects[ob.object_property] = [(ob.object_property, location, ob.confidence)]
            else:
                objects[ob.object_property].append((ob.object_property, location, ob.confidence))

    return tags, objects
```

```python
def predicted_classification_of_one_image(image_path):
    tags, objects = get_predicted_class(image_path, bycicle_obj)
    # if len(tags) == 0 and len(objects) == 0:
    #     return "without_bike"
    # else:
    #     return "with_bike"
    # TIN sa precizez ca daca ne uitam numai in objects se schimba totu
    # La imaginea 8 unde se afla bicicleta albastra el o identifica ca fiind o bicicleta
    # insa nu ii identifica pozitia corecta in imagine asa ca nu o pune ca si object
    if len(objects) == 0:
        return "without_bike"
    else:
        return "with_bike"
```

```python
def get_predicted_classes():
    result = {}
    for img in data:
        path = img["image_path"]
        result[path] = predicted_classification_of_one_image(path)
    return result
```

```python
get_predicted_classes()
```

```python
def evaluate_model():
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    predicted = get_predicted_classes()
    for img in data:
        path = img["image_path"]
        if img["type"] == "with_bike":
            if predicted[path] == "with_bike":
                TP += 1
            else:
                FN += 1
        else:
            if predicted[path] == "without_bike":
                TN += 1
            else:
                FP += 1
    precision = TP / (TP + FP) if TP + FP != 0 else 0
    recall = TP / (TP + FN) if TP + FN != 0 else 0
    f1_score = 2 * (precision * recall) / (precision + recall) if precision + recall != 0 else 0
    return precision, recall, f1_score
```

```python
evaluate_model()
```

## Measuring the performance of the algorithm through Intersection Over Union

```python
def calculate_iou(box1,box2):
    #iou= i/union else 0
    #calculam intersectia
    x1=max(box1[0],box2[0])
    y1=max(box1[1],box2[1])
    x2=min(box1[2],box2[2])
    y2=min(box1[3],box2[3])
    area_of_intersection=max(0,x2-x1)*max(0,y2-y1)
    #union=area1+are2-inters
    area_box1=(box1[2]-box1[0]+1)*(box1[3]-box1[1]+1)
    area_box2=(box2[2]-box2[0]+1)*(box2[3]-box2[1]+1)
    union=area_box1+area_box2-area_of_intersection
    #calculam iou
    if union==0:
        return 0
    else:
        return area_of_intersection/union
```

## Mean Squared Measure to identify performance

```python
def calculate_mes_img(pred,gt):
    err=0
    total_mse=0
    num_boxes=0
    i=0
    j=0
    if(len(pred)==0):
        return 1000,len(pred)
    while i<len(pred) and j<len(gt):
        if(abs(pred[i][0]-gt[j][0])>25):
            #inseamna ca nu sunt acelasi boxuri
            err+=100
            j+=1
        else:
            err=sum((pred-gt)**2 for pred,gt in zip(pred[i],gt[j]))/4
            i+=1
            j+=1
        total_mse+=err
        num_boxes+=1
    if num_boxes>0:
        return total_mse/num_boxes,num_boxes
```

```python
def calculate_mes():
    predicted = get_predicted_locations()
    ground_truth = get_ground_truth_locations()
    total_mse = 0
    num_boxes = 0  # Counter for identified boxes

    for pred_boxes, gt_boxes in zip(predicted, ground_truth):
        mse, num_boxes_img = calculate_mes_img(pred_boxes, gt_boxes)
        total_mse += mse
        num_boxes += num_boxes_img


    # Avoid division by zero
    if num_boxes > 0:
        return total_mse / num_boxes
    else:
        return float('inf')  # Return infinity if no identified boxes
```

```python
calculate_mes()
```

```python
def get_nr_of_bikes():
    result = []
    for img in data:
        if img["type"]=="with_bike":
            result.append(img["nr_of_bikes"])
    return sum(result)
```

# Iou overall

```python
def calculate_iou_overall():
    full_iou=0
    for img in data:
        if img["type"]=="with_bike":
            image_path=img["image_path"]
            _,objects=get_predicted_class(image_path,bycicle_obj)
            iou=0
            loc_p=[obj[1] for obj_list in objects.values() for obj in obj_list]
            loc=[loc for loc in img["locations"].values()]
            i=0
            j=0
            while i<len(loc_p) and j<len(loc):
                if(calculate_iou(loc_p[i],loc[j])<0.2):
                    #inseamna ca aici e missing
                    iou+=0
                    j+=1
                else:
                    iou+=calculate_iou(loc_p[i],loc[j])
                    i+=1
                    j+=1
            iou/=len(loc)
            full_iou+=iou
    return full_iou/get_nr_of_bikes()
```

```python
calculate_iou_overall()
```
